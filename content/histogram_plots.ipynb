```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcdefaults()  # reset to default
plt.style.use('phys-plots.mplstyle')
```

```python
# Histogram 1
bins1 = [350, 400, 450, 500, 550, 600, 650, 700]
values1 = [0, 1, 9, 6, 2, 2, 0]

# Histogram 2
bins2 = [0, 50, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]
values2 = [0, 0, 6, 22, 18, 5, 5, 2, 0, 0, 0]

# Plot
scale = 0.65
plt.figure(
    figsize=(3.0*scale, 2.5*scale)
)
plt.hist(bins2[:-1], bins=bins2, weights=values2, alpha=1, label='Hist 2', edgecolor='none', color='gray')
plt.hist(bins1[:-1], bins=bins1, weights=values1, alpha=1, label='Hist 1', edgecolor='none', color='blue')

plt.xlabel('Size (nm)')
plt.ylabel('Counts')
plt.xlim(0, 800)

plt.savefig('results/hist_size_counts.pdf')

plt.show()
```
